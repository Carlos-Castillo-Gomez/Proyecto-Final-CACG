```python
#Se define el cuartil 75 y se le resta el cuartil 25
iqr=df['Cilindros'].quantile(0.75)-df['Cilindros'].quantile(0.25)
#Desarrollamos filtros superior o inferior por lo general es el iqr por 1.5 pero se puede manejar el número
filtro_inferior = df['Cilindros'] > df['Cilindros'].quantile(0.25) - (iqr * 1.5)
filtro_superior = df['Cilindros'] < df['Cilindros'].quantile(0.75) + (iqr * 1.5)

df_filtrado = df[filtro_inferior & filtro_superior]

#Graficando el boxplot
fig = px.box(df_filtrado, y='Cilindros', title='Datos atípicos eliminados de Cilindros')
fig.update_layout(width=300,height=500)
fig.show()
```
