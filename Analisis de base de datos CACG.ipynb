```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Crear datos de ejemplo
X = df2[['Cilindros','Potencia (HP)','Tamaño (L)']]
# Generar una variable objetivo con algo de ruido
y = df2['CO2(g/km)']

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Crear y entrenar el modelo de regresión lineal múltiple
model = LinearRegression()
model.fit(X_train, y_train)

# Hacer predicciones
y_pred = model.predict(X_test)


# Evaluar el modelo
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# Graficar resultados
plt.figure(figsize=(10, 5))
plt.scatter(y_test, y_pred, color="blue", edgecolor="k", s=50, alpha=0.7)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', lw=2)
plt.xlabel("Valores Reales")
plt.ylabel("Predicciones")
plt.title("Regresión Lineal Múltiple")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset (puedes cargarlo desde un archivo CSV)
df = pd.read_csv('Base_limpia_proyecto_CACG')  # Cambia 'ruta_del_archivo.csv' por el archivo que desees

# Supongamos que las columnas a comparar se llaman 'columna1' y 'columna2'

# Crear un DataFrame con las dos columnas que deseas comparar
df_comparar = df[['Cilindros', 'Tamaño (L)']]

# Usamos melt para reestructurar el DataFrame
df_comparar_melt = df_comparar.melt(var_name='Columna', value_name='Valor')

# Crear el gráfico de violín
plt.figure(figsize=(10, 6))  # Tamaño del gráfico

sns.violinplot(x='Columna', y='Valor', data=df_comparar_melt, inner="quart", palette="muted")

# Título y mostrar el gráfico
plt.title('Comparación de Distribución de las Columnas con Gráficos de Violín')
plt.show()
```
